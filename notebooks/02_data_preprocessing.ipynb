# Data Preprocessing and Feature Engineering

```python
import pandas as pd
import numpy as np

from  sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("../data/sample/unsw_nb15_sample.csv")
df.shape
```

```python
attack_labels = df["attack_cat"]

df_model = df.drop(columns=["attack_cat"], errors="ignore")
```

```python
X = df.drop(columns=["label"])
y = df["label"]

X.shape, y.shape
```

```python
cat_cols = X.select_dtypes(include=["object", "string"]).columns
cat_cols
```

```python
le = LabelEncoder()

for col in cat_cols:
    X[col] = le.fit_transform(X[col])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
processed_df = pd.DataFrame(X_scaled, columns=X.columns)
processed_df["label"] = y.values

processed_df.to_csv("../data/processed/unsw_nb15_processed.csv", index=False)
```
